```python
bike.ex_showroom_price.fillna(bike.ex_showroom_price.mode()[0],inplace=True)
```

```python
mode_value = bike.ex_showroom_price.mode()[0]
bike['ex_showroom_price'] = bike['ex_showroom_price'].fillna(mode_value)
```

```python
bike.isnull().sum()
```

```python
bike.replace({'owner':{'1st owner':0,'2nd owner':1,'3rd owner':2,'4th owner':3}},inplace=True) 
bike.replace({'seller_type':{'Dealer':0,'Individual':1}},inplace=True) 
bike
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X = bike.drop(['name','selling_price'],axis=1) 
Y = bike['selling_price'] 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42) 
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
lrmodel = LinearRegression()  
lrmodel.fit(X_train,Y_train)
```

```python
import joblib
joblib.dump(lrmodel, 'motorcycle_price_model.pkl')
```
